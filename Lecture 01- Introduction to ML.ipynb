## Supervised Learning Examples:

```python
import pandas as pd

house_data = {
    "Area": [1000,1200,1500],
    "Price": [20,24,30]
}

df = pd.DataFrame(house_data)

df
```

Features:
- Area
- Bedrooms

Label:
- Price

Problem Type:
- Regression

```python
student_data = {
    "Study Hours": [1, 2, 3, 7, 8, 9],
    "Result": ["Fail", "Fail", "Fail", "Pass", "Pass", "Pass"]
}

students = pd.DataFrame(student_data)

students
```

Feature:
- Study Hours

Label:
- Result

Problem Type:
- Classification

# Unsupervised Learning Example:

```python
customer_data = {
    "Annual Income": [20000, 25000, 100000, 120000],
    "Spending Score": [20, 25, 90, 95]
}

customers = pd.DataFrame(customer_data)

customers
```

No labels are available.

The goal is to discover hidden groups of customers.

Problem Type:
- Unsupervised Learning

# Reinforcement Learning Example:

```python
reward = 10

if reward > 0:
    print("Good Action -> Reward Received")
else:
    print("Bad Action -> Penalty")
```

Examples:
- Self-Driving Cars
- Robotics
- Game Playing AI

The agent learns using rewards and penalties.

### Key Learnings from Lecture 1

1. Machine Learning learns patterns from data.
2. Supervised Learning uses labeled data.
3. Regression predicts numerical values.
4. Classification predicts categories.
5. Unsupervised Learning discovers hidden patterns.
6. Reinforcement Learning learns through rewards and penalties.

Most ML projects start by identifying:
- Features
- Labels
- Problem Type
